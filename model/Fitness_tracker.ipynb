```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('sample_data/fitness.csv')
df.head()
```

```python
df1 = df.copy()
for col in df1.columns:
    if df1[col].dtype == object:
      dv = pd.get_dummies(df[col],drop_first=True)
      df1 = pd.concat([df1,dv],axis=1)
      df1.drop(col,axis=1,inplace=True)
df1.info()
```

```python
df1.drop("Resting Heart Rate (bpm)",axis=1,inplace=True)
df1.info()
```

```python
df1['BMI'] = df["Weight (kg)"] / (df["Height (cm)"] / 100)**2

df1.drop(columns=["Height (cm)", "Weight (kg)"], inplace=True)

df1.head()
```

```python
from sklearn.cluster import KMeans
km = KMeans()
df1["cluster"] = km.fit_predict(df1)
df1.head()
df1.describe()
```

```python
k_rng = range(1,11)
sse = []
for i in k_rng:
  km = KMeans(n_clusters=i)
  km.fit(df1)
  sse.append(km.inertia_)

plt.plot(k_rng,sse)
plt.xlabel("K")
plt.ylabel("SSE")
plt.show()
```

```python

df2 = df1.copy()
df2.drop("cluster",axis=1,inplace=True)
optimal_clusters = 2
km = KMeans(n_clusters=optimal_clusters,random_state=0)
df2["cluster"] = km.fit_predict(df2)
df2.head(10)
```

```python
cluster_means = df2.groupby("cluster").mean()
cluster_means
```

```python
df2.info()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
#dropped more unecessary columns
df2.drop(["User ID","Calories Burned","Heart Rate (bpm)","Distance (km)","VO2 Max","Body Fat (%)"],axis=1,inplace=True)
features = df2.drop(["cluster"],axis=1)
corr_matrix = features.corr().abs()
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
threshold = 0.2
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > threshold)]

df3 = features.drop(columns=to_drop)

print("Dropped features:", to_drop)
print("Remaining features:", df3.columns)
df3.info()
```

```python
X = df3
y = df2["cluster"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVC(random_state=42)
model.fit(X_train, y_train)
df3.info()
y_pred = model.predict(X_test)
print(model.score(X_test,y_test))
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import pickle
with open("fitness_model.pkl","wb") as f:
  pickle.dump(model,f)
```
